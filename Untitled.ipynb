```python
from illuminating.data_preparation.geodata_processing import download_gdf
```

```python
USA_gdf = download_gdf("USA",2)
```

```python
Arizona_gdf =USA_gdf[USA_gdf.NAME_1=="Arizona"]
Arizona_gdf.head()
```

```python
minx,miny,maxx,maxy = Arizona_gdf.total_bounds
```

```python
import geopandas as gpd
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:lewagon@localhost/geospatial_db')
query = f"""
SELECT *
FROM road
WHERE geometry && ST_MakeEnvelope({minx}, {miny}, {maxx}, {maxy}, 4326);
"""
gdf_bbox = gpd.read_postgis(query, engine, geom_col='geometry')
```

```python
gdf_bbox.head()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 10))
Arizona_gdf.boundary.plot(ax=ax, color='black', linewidth=2)
gdf_bbox.plot(ax=ax,color='green',linewidth=0.5)
ax.set_xlim([minx, maxx])
ax.set_ylim([miny, maxy])
plt.show();
```
