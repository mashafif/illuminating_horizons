```python
import os
import geopandas as gpd
import pandas as pd
import glob

def download_gdf(state_name: str, country_code: str = "USA", resolution: int = 1):
    base_url = "https://geodata.ucdavis.edu/gadm/gadm4.1/json/gadm41"
    download_url = f"{base_url}_{country_code}_{resolution}.json.zip"
    gdf = gpd.read_file(download_url)

    # Filter the GeoDataFrame to the specific state
    gdf_state = gdf[gdf['NAME_1'] == state_name]

    return gdf_state

def download_all_us_states_gdfs(resolution: int = 1, save_path: str = "gadm_gdf"):
    # Create a directory to save the GeoDataFrames
    os.makedirs(save_path, exist_ok=True)
    # List of U.S. states with their corresponding abbreviations
    us_states = {
        'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
        'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
        'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
        'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
        'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
        'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
        'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
        'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
        'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
        'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
    }
    for state_name, state_abbr in us_states.items():
        if os.path.join(save_path, f"gdf_US-{state_abbr}.geojson") not in glob.glob('./gadm_gdf/*.geojson'):
            try:
                print(f"Downloading GeoDataFrame for {state_name} ({state_abbr})...")
                gdf_state = download_gdf(state_name=state_name.replace(' ',''), country_code="USA", resolution=resolution)

                if gdf_state is not None and not gdf_state.empty:
                    # Save the GeoDataFrame to a file, using the state abbreviation in the filename
                    output_file = os.path.join(save_path, f"gdf_US-{state_abbr}.geojson")
                    gdf_state.to_file(output_file, driver="GeoJSON")
                    print(f"Saved: {output_file}")

            except Exception as e:
                print(f"Failed to download GeoDataFrame for {state_name}: {e}")

    print("Download complete.")

# Example usage
download_all_us_states_gdfs(resolution=1, save_path="gadm_gdf")
```

```python
import os
import geopandas as gpd
import pandas as pd

def combine_state_geojsons(input_folder: str, output_file: str):
    # List all GeoJSON files in the input folder
    geojson_files = [f for f in os.listdir(input_folder) if f.endswith('.geojson')]
    #geojson_files = [x.split('/')[-1] for x in glob.glob(input_folder+'/*_US-*.geojson')]
    print(geojson_files)
    # List to store GeoDataFrames
    gdfs = []

    # Loop through each GeoJSON file and append it to the list
    for geojson_file in geojson_files:
        if "US-AK" in geojson_file or "US-HI" in geojson_file:
            print(f"Skipping: {geojson_file}")  # Skip Alaska and Hawaii
            continue

        print(f"Processing: {geojson_file}")
        #gdf = gpd.read_file(os.path.join(input_folder, geojson_file))
        gdf = gpd.read_file(input_folder+'/'+geojson_file)
        gdfs.append(gdf)

    # Concatenate all GeoDataFrames
    conus_gdf = pd.concat(gdfs, ignore_index=True)

    # Dissolve the GeoDataFrame to get the outer boundary (convex hull)
    #conus_boundary = conus_gdf.unary_union.convex_hull

    # Create a new GeoDataFrame with the boundary
    boundary_gdf = gpd.GeoDataFrame(geometry=[conus_gdf], crs="EPSG:4326")

    # Save the combined GeoDataFrame as a single GeoJSON file
    boundary_gdf.to_file(output_file, driver="GeoJSON")
    print(f"Saved combined CONUS GeoDataFrame to {output_file}")

# Example usage
combine_state_geojsons(input_folder="../raw_data/gadm_gdf", output_file="gdf_CONUS_raw.geojson")
```

```python
import xarray as xr
import os
import numpy as np

# Define the directory where the NetCDF files are stored
nc_dir = '../raw_data/bm/'  # Update this path to where your files are stored


# List all NetCDF files in the directory
nc_files = [os.path.join(nc_dir, file) for file in os.listdir(nc_dir) if file.endswith('.nc')]

# Filter out Alaska and Hawaii
nc_files = [file for file in nc_files if not any(state in file for state in ['US-AK', 'US-HI'])]

# Initialize an empty list to store datasets
datasets = []

for file in nc_files:
    ds = xr.open_dataset(file)
    datasets.append(ds)

# Combine all datasets into one, with coords='minimal'
combined_ds = xr.concat(datasets, dim='state', combine_attrs='drop_conflicts', coords='minimal')

# Convert data types to reduce size (use float16)
combined_ds = combined_ds.astype('float16')

# Save the combined dataset with compression using the netCDF4 engine
comp = dict(zlib=True, complevel=5)
encoding = {var: comp for var in combined_ds.data_vars}

# Correct save path
output_path = '../raw_data/CONUS.nc'  # Ensure this path is correct and writable

# Create the directory if it doesn't exist
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Save the combined dataset
combined_ds.to_netcdf(output_path, encoding=encoding, engine='netcdf4')
```
